```python
#transforming data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train_data_trans = sc.fit_transform(train_data)
test_data_trans = sc.transform(test_data)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from sklearn.feature_selection import f_classif
def select_features(train_data_trans, train_label, test_data_trans):
  # configure to select a subset of features
  fs = SelectKBest(score_func=f_regression, k=20)
  fsc = SelectKBest(score_func=f_classif, k=20)
  # learn relationship from training data
  fs.fit(train_data_trans, train_label)
  fsc.fit(train_data_trans, train_label)
  # transform train input data
  train_fs = fs.transform(train_data_trans)
  train_fsc = fsc.transform(train_data_trans)
  # transform test input data
  test_fs = fs.transform(test_data_trans)
  test_fsc = fsc.transform(test_data_trans)
  #X_new = selector.fit_transform(X, y)
  names = train_data.columns.values[fs.get_support()]
  namesc = train_data.columns.values[fsc.get_support()]
  scores = fs.scores_[fs.get_support()]
  scoresc = fsc.scores_[fsc.get_support()]
  names_scores = list(zip(names, scores))
  names_scoresc = list(zip(namesc,scoresc))
  ns_df = pd.DataFrame(data = names_scores, columns=['Feat_names', 'F_Scores'])
  ns_dfc = pd.DataFrame(data = names_scoresc, columns=['Feat_names', 'F_Scores'])
  #Sort the dataframe for better visualization
  ns_df_sorted = ns_df.sort_values(['F_Scores', 'Feat_names'], ascending = [False, True])
  ns_df_sortedc = ns_dfc.sort_values(['F_Scores', 'Feat_names'], ascending = [False, True])
  print(ns_df_sorted)
  print(ns_df_sortedc)
  return train_fs, test_fs, fs, train_fsc, test_fsc, fsc, ns_df_sorted, ns_df_sortedc
```

```python
train_fs, test_fs, fs, train_fsc, test_fsc, fsc, ns_df_sorted, ns_df_sortedc = select_features(train_data_trans, train_label, test_data_trans)
print(train_fsc.shape)
print(train_fs.shape)
print(train_label.shape)
print(test_fsc.shape)
print(test_fs.shape)
print(test_label.shape)
ns_df_sortedc["Feat_names"]
```

### LOGISTIC REGRESSION

```python
def calc_FNR(cnf_matrix):
  FP = cnf_matrix.sum(axis=0) - np.diag(cnf_matrix)  
  FN = cnf_matrix.sum(axis=1) - np.diag(cnf_matrix)
  TP = np.diag(cnf_matrix)
  TN = cnf_matrix.sum() - (FP + FN + TP)
  FP = FP.astype(float)
  FN = FN.astype(float)
  TP = TP.astype(float)
  TN = TN.astype(float)
  # Sensitivity, hit rate, recall, or true positive rate
  TPR = TP/(TP+FN)
  # Specificity or true negative rate
  TNR = TN/(TN+FP) 
  # Precision or positive predictive value
  PPV = TP/(TP+FP)
  # Negative predictive value
  NPV = TN/(TN+FN)
  # Fall out or false positive rate
  FPR = FP/(FP+TN)
  # False negative rate
  FNR = FN/(TP+FN)
  print("false negative rate", FNR)
  # False discovery rate
  FDR = FP/(TP+FP)
  # Overall accuracy
  ACC = (TP+TN)/(TP+FP+FN+TN)
  
```
